```python
import librosa
import librosa.display
```

```python
audio_file = "9.wav"
y, sr = librosa.load(audio_file)
```

```python
window_size = 1024
window = np.hanning(window_size)
```

```python
y_windowed = window * y[:window_size]
X = fft(y_windowed)
magnitude = np.abs(X)
log_magnitude = np.log(magnitude + np.spacing(1))
cepstrum = ifft(log_magnitude)
cepstrum = cepstrum.real
```

```python
cepstrum /= np.max(np.abs(cepstrum))

# Plot cepstrum
plt.figure(figsize=(10, 5))
plt.plot(cepstrum)
plt.title('Cepstrum')
plt.xlabel('Time (samples)')
plt.ylabel('Cepstral Coefficient (Normalized)')
plt.tight_layout()
plt.show()
```
