```python
import librosa
import numpy as np
import os
import numpy as np
import IPython.display as ipd
import pywt
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt
```

```python
def compute_wavelet_packets(signal, wavelet, level):
    coeffs = pywt.wavedec(signal, wavelet, level=level)
    return coeffs
```

```python
sample_rate, audio_data = wavfile.read('31.wav')

if len(audio_data.shape) > 1:
    audio_data = audio_data[:, 0]
wavelet = 'db4'
coeffs = compute_wavelet_packets(audio_data, wavelet, 5)
```

```python
plt.figure(figsize=(10, 6))
for i in range(6):
    plt.subplot(6, 1, i + 1)
    plt.plot(coeffs[i])
    plt.title(f'Level {i} Coefficients')
plt.tight_layout()
plt.show()
```
